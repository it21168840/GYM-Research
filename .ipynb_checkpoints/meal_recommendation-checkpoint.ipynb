```python
import json
import pickle
import pprint
import numpy as np 
import pandas as pd
import seaborn as sns
from collections import defaultdict
from matplotlib import pyplot as plt
from catboost import CatBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
def data_pipeline(datapath = 'data/UserList50.xlsx'):
    df = pd.read_excel(datapath)
    print(df.isnull().sum() / len(df))
    del df['ID'], df['Allergies'], df['Health Condition']
    df = df.dropna()

    cat_columns = [col for col in df.columns if df[col].dtype == 'object']
    encoder = defaultdict(LabelEncoder)
    for col in cat_columns:
        encoder[col].fit(df[col])
        df[col] = encoder[col].transform(df[col])

    with open('artifacts/encoder_meal.pkl', 'wb') as f:
        pickle.dump(encoder, f)

    Y = df['MP'].values
    X = df.drop('MP', axis=1).values
    return X, Y, encoder
```

```python
X, Y, encoder_meal = data_pipeline()

X_train, X_test, Y_train, Y_test = train_test_split(
                                                    X, Y, 
                                                    test_size=0.2, 
                                                    random_state=42
                                                    )

print(f"\nX_train shape: {X_train.shape}")
print(f"Y_train shape: {Y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Y_test shape: {Y_test.shape}")
```

```python
cat = CatBoostClassifier(
                        iterations=1000, 
                        learning_rate=0.1, 
                        loss_function='MultiClass', 
                        depth=6
                        )
cat.fit(
        X, Y, 
        eval_set=(
                    X_test, 
                    Y_test
                ), 
        verbose=100
        )
```

```python
with open('artifacts/model_meal.pkl', 'wb') as f:
    pickle.dump(cat, f)
```

```python
P_train = cat.predict(X_train)
P_test = cat.predict(X_test)
```

```python
print("---------------------- Train CLS REPORT ----------------------")
clf_report = classification_report(
                                Y_train,
                                P_train,
                                target_names = encoder_meal['MP'].classes_
                )
print(clf_report)

print("---------------------- Test CLS REPORT ----------------------")
clf_report = classification_report(
                                Y_test, 
                                P_test,
                                target_names = encoder_meal['MP'].classes_
                                )
print(clf_report)
```

```python
cm = confusion_matrix(Y_train,P_train)
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
cm = np.round(cm, 2)

plt.figure(figsize=(40, 40))
sns.heatmap(cm, annot=True, fmt=".2f", cmap='Blues', xticklabels=encoder_meal['MP'].classes_, yticklabels=encoder_meal['MP'].classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Train Confusion Matrix')
plt.show()
```

```python
cm = confusion_matrix(Y_test,P_test)
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
cm = np.round(cm, 2)

plt.figure(figsize=(40, 40))
sns.heatmap(cm, annot=True, fmt=".2f", cmap='Blues', xticklabels=encoder_meal['MP'].classes_, yticklabels=encoder_meal['MP'].classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Test Confusion Matrix')
plt.show()
```

### Inference

```python
with open('artifacts/encoder_meal.pkl', 'rb') as f:
    encoder_meal = pickle.load(f)

with open('artifacts/model_meal.pkl', 'rb') as f:
    model_meal = pickle.load(f)
```

```python
def inference_meal(
                    sample_json,
                    meal_path = 'data/Plans_ID50.xlsx'
                    ):
    df = pd.DataFrame([sample_json])
    del df['ID'], df['Allergies'], df['Health Condition']

    cat_columns = [col for col in df.columns if df[col].dtype == 'object']
    for col in cat_columns:
        df[col] = encoder_meal[col].transform(df[col])

    X = df.values
    Y = model_meal.predict(X)
    P = encoder_meal['MP'].inverse_transform(Y)
    P = str(P[0])

    df_meal = pd.read_excel(meal_path)
    df_meal = df_meal[df_meal['ID'].str.contains(P)]
    df_meal = df_meal.to_dict(orient='records')
    return df_meal
```

```python
sample_json = {
            "ID":"U0002",
            "Fitness Goal":"Maintenance",
            "Allergies":"Seafoods",
            "Veganism":"Non-veg",
            "Health Condition":"Diabetes",
            "Preferences":"Mix",
            "Activity Level":"Very active",
            "Body Type":"Lean",
            "Workout Routine":"Weight training",
            }
```

```python
df_meal = inference_meal(sample_json)
pprint.pprint(df_meal)
```
